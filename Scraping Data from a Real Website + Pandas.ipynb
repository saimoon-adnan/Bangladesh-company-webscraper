```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from bs4.element import Tag
```

```python
URL = "https://en.wikipedia.org/wiki/List_of_companies_of_Bangladesh"
```

```python
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})
```

```python
webpage = requests.get (URL,headers=HEADERS)
```

```python
type(webpage.content)
```

```python
#soup object containiang all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
#assume this table class is a wikitable
tables = soup.find("table", {"class": "wikitable"})
```

```python
tables = [t for t in soup.find_all("table") if isinstance(t, Tag)]

dfs = []

for table in tables:
    if "wikitable" not in table.get("class", []):
        continue

    rows = table.find_all("tr")

    # Headers
    headers = [th.text.strip() for th in rows[0].find_all("th")]

    data = []
    for row in rows[1:]:
        cols = [td.text.strip() for td in row.find_all(["td","th"])]

      
        if len(cols) > len(headers):
            cols = cols[:len(headers)]

    
        if len(cols) < len(headers):
            cols.extend([""] * (len(headers) - len(cols)))

        data.append(cols)

    df = pd.DataFrame(data, columns=headers)
    dfs.append(df)

dfs[0]
```

```python
import os

# Target folder
save_path = r"D:\JUPYTER NOTEBOOK PROJECT"

# If folder does not exist, create it (optional)
os.makedirs(save_path, exist_ok=True)

# File name
file_name = "bangladesh_companies.csv"

# Full path
full_path = os.path.join(save_path, file_name)

# Save CSV
df.to_csv(full_path, index=False)

print("CSV saved successfully at:", full_path)
```
